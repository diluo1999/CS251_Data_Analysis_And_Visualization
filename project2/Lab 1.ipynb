Visualize the dataset with Matplotlib. All charts should include a title, axis labels and a legend when applicable.
1. Import 'lab1.csv' into a ndarray using genfromtxt. 
2. Set figure size to 12x8. [Figure reference](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.figure.html)
2. Create a cartesian line graph of column 1 (on the X axis) vs column 0 (Y axis) using a magenta dashed line.
    1. Include a title and axis labels.
3. Create a cartesian scatterplot with 3 datasets, (1,0) as red stars, (2,1) as green dots and (2,0) as blue squares.
    1. Include a title, and a legend.
4. Create 6 graphs in 2 columns and 3 rows using [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html). 
    1. Row 1 is two graphs of (1,0), the first as a scatterplot and the second as a line graph
    2. Row 2 is two graphs of (2,1), the first as a scatterplot and the second as a line graph
    3. Row 3 is two graphs of (2,0), the first as a scatterplot and the second as a line graph
    4. Include a title and axis labels for each subplot.
    5. To insure there is enough space between subplots use [subplots_adjust](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots_adjust.html)

```python
#part 2 here
lab1=np.genfromtxt('data/lab1.csv',delimiter=',')
print(lab1)
print(lab1.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
x=lab1[:,1]
y=lab1[:,0]
plt.plot(x,y,linestyle='--',color='magenta')
plt.xlabel('X Axis: Column 1')
plt.ylabel('Y Axis: Column 0')
plt.title('3. column 1 vs column 0')
plt.show()
```

```python
plt.scatter(lab1[:,1], lab1[:,0], c='red', marker=(5, 1), label='(1,0)')
plt.scatter(lab1[:,2], lab1[:,1], c='green', label='(2,1)')
plt.scatter(lab1[:,2], lab1[:,0], c='blue', marker='s', label='(2,0)')
plt.title('4. Scatter plot')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(5,10))
plt.subplots_adjust(wspace=1, hspace=1)
axes[0,0].scatter(lab1[:,1], lab1[:,0])
axes[0,0].set_title('(1,0)scatter')
axes[0,0].set_xlabel('Column 1')
axes[0,0].set_ylabel('Column 0')
axes[0,1].plot(lab1[:,1], lab1[:,0])
axes[0,1].set_title('(1,0)line')
axes[0,1].set_xlabel('Column 1')
axes[0,1].set_ylabel('Column 0')

axes[1,0].scatter(lab1[:,2], lab1[:,1])
axes[1,0].set_title('(2,1)scatter')
axes[1,0].set_xlabel('Column 2')
axes[1,0].set_ylabel('Column 1')
axes[1,1].plot(lab1[:,2], lab1[:,1])
axes[1,1].set_title('(2,1)scatter')
axes[1,1].set_xlabel('Column 2')
axes[1,1].set_ylabel('Column 1')

axes[2,0].scatter(lab1[:,2], lab1[:,0])
axes[2,0].set_title('(2,0)scatter')
axes[2,0].set_xlabel('Column 2')
axes[2,0].set_ylabel('Column 0')
axes[2,1].plot(lab1[:,2], lab1[:,0])
axes[2,1].set_title('(2,0)scatter')
axes[2,1].set_xlabel('Column 2')
axes[2,1].set_ylabel('Column 0')

print('(1,0) and (2,0) are better as scatter plots since their data are more random.\n')
print('(2,1) is better as line graphs since the data are continuous and regular.\n')
```

#### Using the above links and MatPlotLib documentation to complete the following tasks:

Visualize the dataset with Matplotlib. All charts must include a title and axis labels.

1. A 3D Cartesian figure: col 2 (X), col 1 (Y), and col 0 (Z). 
2. A figure with 6 subplots
    1. First row histograms with 10 bins of col 0, col 1, and col 2.
    1. Second row scatter or line graphs (whichever you think is more meaningful) of (0,2), (1,2), and (2,2).  

```python
##3d figure here
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=lab1[:,2], ys=lab1[:,1], zs=lab1[:,0])
ax.set_title('3D Cartesian Figure')
ax.set_xlabel('X: col2')
ax.set_ylabel('Y: col1')
ax.set_zlabel('Z: col0')

plt.show()
```

```python
##Histograms here
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8,10), tight_layout=True)
plt.subplots_adjust(wspace=5, hspace=5)
axes[0,0].hist(x=lab1[:,0], bins=10)
axes[0,0].set_title('col0 histograms')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')
axes[0,1].hist(x=lab1[:,1], bins=10)
axes[0,1].set_title('col1 histograms')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')
axes[0,2].hist(x=lab1[:,2], bins=10)
axes[0,2].set_title('col2 histograms')
axes[0,2].set_xlabel('x')
axes[0,2].set_ylabel('y')

axes[1,0].plot(lab1[:,0], lab1[:,2])
axes[1,0].set_title('(0,2)plot')
axes[1,0].set_xlabel('Column 0')
axes[1,0].set_ylabel('Column 2')
axes[1,1].plot(lab1[:,1], lab1[:,2])
axes[1,1].set_title('(1,2)plot')
axes[1,1].set_xlabel('Column 1')
axes[1,1].set_ylabel('Column 2')
axes[1,2].plot(lab1[:,2], lab1[:,2])
axes[1,2].set_title('(2,2)plot')
axes[1,2].set_xlabel('Column 2')
axes[1,2].set_ylabel('Column 2')
```
