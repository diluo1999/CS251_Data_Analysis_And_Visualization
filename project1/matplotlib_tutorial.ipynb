```python
import numpy as np
# import matplotlib plotting library
import matplotlib.pyplot as plt
```

## Basic plotting example

### Plot log(x)

```python
# make a line plot: generate x 50 values evenly spaced between +1 and +50
x = np.linspace(1, 50)
# Evaluate the log
y = np.log(x)
# plot it!
plt.plot(x, y)
plt.show()
```

### Plot multiple log curves in same plot

```python
# We can plot a few more on the same plot
y_2 = np.log(20*x)
y_3 = np.log(30*x)

# plot them!
plt.plot(x, y)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.show()
```

### Customize colors

```python
# plot them!
plt.plot(x, y, 'b')
plt.plot(x, y_2, 'k')
plt.plot(x, y_3, 'g')
plt.show()
```

### Use colorbrewer qualitative color palette

Options here: https://github.com/CartoDB/CartoColor/wiki/CARTOColor-Scheme-Names

```python
from palettable import cartocolors
```

```python
# Define safe color scheme, specify RGB color with color optional parameter
brewer_colors = cartocolors.qualitative.Safe_3.mpl_colors

# plot them!
plt.plot(x, y, color=brewer_colors[0])
plt.plot(x, y_2, color=brewer_colors[1])
plt.plot(x, y_3, color=brewer_colors[2])
plt.show()
```

### Customize examples: line thickness and style

```python
plt.plot(x, y, color=brewer_colors[0], linewidth=3, linestyle='dashed')
plt.plot(x, y_2, color=brewer_colors[1], linewidth=5, linestyle='dotted')
plt.plot(x, y_3, color=brewer_colors[2], linewidth=7)
plt.show()
```

### Set title, axis viewing ranges, legend

```python
plt.plot(x, y, color=brewer_colors[0], linewidth=3, linestyle='dashed', label='log1')
plt.plot(x, y_2, color=brewer_colors[1], linewidth=5, linestyle='dotted', label='log20')
plt.plot(x, y_3, color=brewer_colors[2], linewidth=7, label='log30')
plt.xlabel('My x axis')
plt.ylabel('My y axis')
plt.xlim([0, 60])
plt.ylim([0, 10])
plt.title('My pretty plot!')
plt.legend()
plt.show()
```

## Subplots

**Matplotlib is powerful!** 

- Let's create a grid of stacked plots with `plt.subplots`.
- Let's make it 2 rows and 1 col. 
- Let's put the log plot on top and a scatter plot of random data below

```python
# The "grid" object of 2 subplots is called a Figure object
# The subplots are called Axes
# Each axis acts like plt above, except the plot decoration commands
# usually have a set_ before them. e.g. ax.set_xlabel('My x label')
# Control figure size with figsize optional parametr of plt.subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5,10))
axes[0].plot(x, y, color=brewer_colors[0], linewidth=3, linestyle='dashed', label='log1')
axes[0].plot(x, y_2, color=brewer_colors[1], linewidth=5, linestyle='dotted', label='log20')
axes[0].plot(x, y_3, color=brewer_colors[2], linewidth=7, label='log30')
# set the title
axes[0].set_title('Log plot')

# Generate random data in both x and y
x_r = np.random.random([50,])
y_r = np.random.random([50,])

# Make scatter plot
axes[1].scatter(x_r, y_r)
axes[1].set_title('Scatter plot')
plt.show()
```
